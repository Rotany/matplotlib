## Plot top 5 poblacion por pais y año

```python
import matplotlib.pyplot as plt

def plot_for_years_and_wanted_column(wanted_column):
    # Obtén los años únicos en tus datos
    years = pai['Year'].unique()

    # Ordena los años en orden ascendente
    years = sorted(years)

    for year in years:
        # Filtra el DataFrame original para cada año
        data_for_year = pai[pai['Year'] == year]

        wanted_column = wanted_column

        # Ordena los datos por la wanted_column
        top_countries = data_for_year.sort_values(wanted_column, ascending=False).head(5)

        # Crea un gráfico de barras para el año actual
        plt.figure(figsize=(10,5))  # Ajusta el tamaño de la figura según sea necesario
        plt.bar(top_countries['Country'], top_countries[wanted_column])

        # Ajusta las etiquetas de los ejes y el título según sea necesario
        plt.xlabel('Country')
        plt.xticks(rotation=45)
        plt.ylabel(wanted_column)
        plt.title(f'Top 5 Countries for {wanted_column} for Year ' + str(year))

        plt.show()

```

## Plot de la esperanza de vida de las mujeres por pais

```python
plot_for_years_and_wanted_column(wanted_column = "Life expectancy(women)")
```

```python
plot_for_years_and_wanted_column("Population")
```
