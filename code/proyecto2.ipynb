```python
years= pai['Year'].unique()
years= sorted(years)


```

```python
data_for_Year=pai[pai['Year']==years]
```

```python
top_country= data_for_Year.sort_values('Adult Mortality (women)',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_country['Country'],top_country['Adult Mortality (women)'])
```

```python
top_Country_Population=pai.groupby('Country') ['Adult Mortality (women)']. sum().sort_values(ascending=False).head(5)
plt.xticks (np.arange(5),("India","Japon","Mexico","Colombia","Alemania"),rotation=45) 
plt.bar(top_Country_Population.index,top_Country_Population , color=[ 'red' , 'blue'])
plt.xticks(rotation=45)
plt.title('Mortalidad de Mujeres por Pais')
plt.ylabel('Adult Mortality (women)')
plt.show()
```

```python
mortalidad_m=pai.groupby(['Adult Mortality (women)','Year']) .max() .head(5)

mortalidad_m
```

```python
plt.figure(figsize=(6,6))
plt.bar(mortalidad_m.index,mortalidad_m)
plt.xlabel()

plt.show()
```

```python
top_Country_Population = pai.groupby('Country')['Population'].sum() \
.sort_values(ascending=False).head(10)
    
    #plot
plt.style.use('ggplot') 
plt.bar(top_Country_Population.index,top_Country_Population)
plt.xticks(rotation=45) 
plt.title('Country Population') 
plt.ylabel('Population')
plt.show()
```

```python
top_Country_Population = pai.groupby('Country')['Population'].sum() .sort_values(ascending=False) .head(10)
    
    #plot
plt.style.use('ggplot') 
plt.bar(top_Country_Population.index,top_Country_Population , color=['lightcoral','rosybrown','brown','red','maroon'])
plt.xticks(rotation=45) 
plt.title('Country Population') 
plt.ylabel('Population')
plt.show()
```
