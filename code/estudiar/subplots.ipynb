```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace (0,5,11)
y = np.sin(x)
x
```

```python
y
```

```python
#para crear una grafica tipo fig.necesitas trabajar el siguiente codigo
```

```python
fig, axes= plt.subplots()
axes.plot(x,y,'b')
plt.show()
```

```python
# varias graficas tipo Fig.
```

```python
fig, axes= plt.subplots (nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

```python
# acceder a los graficos mediante el nombre almacenado en una Tupla
```

```python
fig,(axes1,axes2)=plt.subplots(nrows=1,ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
plt.show()
```

```python
# genera una matriz de tipo fig.con dos filas y 4 columnas
```

```python
fig ,axes= plt.subplots(2,4)
axes[0,0].plot (x,np.cos(x))
axes[0,1].plot (x,np.sin(x),'y')
axes[0,2].plot (x,np.tan(x),'r')
axes[0,3].plot (x,np.cos(x)**2)
```

```python
# alternativa para generar una matriz de tipo fig. accediendo a cada grafica mediante los nombres dados en una Tupla.
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8))= plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
fig.tight_layout()
```

```python
fig ,axes= plt.subplots(2,4)
axes[0,0].plot (x,np.cos(x))
axes[0,1].plot (x,np.sin(x),'y')
axes[0,2].plot (x,np.tan(x),'r')
axes[0,3].plot (x,np.cos(x)**2)
axes[1,0].plot (x,np.cos(x)**4,'r:')
axes[1,1].plot (x,y/5,'bo')
axes[1,2].plot (x,y**4,'g-')
axes[1,3].plot (x*5,y**2,'y<')
plt.show()
fig.tight_layout()
```
