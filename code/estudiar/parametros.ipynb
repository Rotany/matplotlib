```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot (data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='sex', y='smoker')
plt.show()
```

```python
# hacer agrupaciones con la variable 'hue'
```

```python
sns.displot(data=tips, x= 'total_bill',y= 'tip',hue='sex')
plt.show()
```

```python
# cambiar el tipo de grafica con kind
```

```python
sns.displot(data=tips, x='total_bill',hue='sex',legend='True',kind='kde')
plt.show()
```

```python
# quitamos la legenda,cambiamos la paletta de colores y su transparencia
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=False,palette='dark', alpha=0.8)
plt.show()
```
