```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
countrys = ["INDIA", "JAPON", "COLOMBIA", "MEXICO","GERMANY"]
population=[1000,800,900,1000,300]
```

```python
plt.bar (countrys,population)
plt.show()
```

```python
plt.bar (countrys,population,width=0.5 , color= ['red','blue'])
plt.show()
```

```python
plt.bar(countrys,population, width= 0.5, color= ['red','blue'])
plt.xticks (np.arange(5),("India","Japon","Mexico","Colombia","Alemania"),rotation=45)
plt.show()
```

```python
plt.barh(countrys,population,color=['red','blue'])
plt.show()
```

```python
plt.style.use ('grayscale')
plt.bar (countrys,population,width=0.5 , color= ['lightcoral','rosybrown','brown','red','maroon'])
plt.title ('Populations by countries')
plt.xlabel ('Countries')
plt.ylabel('Population',rotation=90)
plt.show()
```

```python
N=50
x =np.random.rand(N)
y =np.random.rand(N)
area = (30*np.random.rand (N))**2
colors= np.random.rand(N)
        
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter (x,y,s=area, c= colors, marker='<',alpha=1)
plt.show()
```
